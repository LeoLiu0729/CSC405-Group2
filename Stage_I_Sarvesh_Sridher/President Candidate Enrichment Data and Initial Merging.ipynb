#Merging Enrichment with Covid-19 Super Set

```python
#Merging Enrichment data with Super Covid-19 dataframe

import pandas as pd

candidate = pd.read_csv('president_county_candidate.csv')
super_covid = pd.read_csv('super_covid_data.csv')

merged_data = pd.merge(candidate, super_covid, left_on=['county'], right_on=['County Name'], how='inner')

merged_data
```

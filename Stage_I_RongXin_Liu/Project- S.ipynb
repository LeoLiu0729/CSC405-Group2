```python
import pandas as pd

cases = pd.read_csv("covid_confirmed_usafacts.csv")
deaths = pd.read_csv("covid_deaths_usafacts.csv")
population = pd.read_csv("covid_county_population_usafacts.csv")
```

```python
pd.merge(cases, deaths, on='countyFIPS', how='inner')
```

```python
covid_df = pd.merge(cases, deaths, on='countyFIPS', how='inner')
```

```python
covid_df = covid_df[covid_df['countyFIPS'] != 0]
```

```python
super_covid_df = pd.merge(covid_df, population, on='countyFIPS', how='inner')
```

```python
super_covid_df.to_csv("super_covid_data.csv", index=False)
```

# Enrichment data

```python
import pandas as pd

enrichment_df = pd.read_csv('State.DP05-2024-01-31T061845.csv')

print(enrichment_df.head())
```

```python
import pandas as pd


state_data = []


for column in enrichment_df.columns:
    if "!!Estimate" in column:
        state_name = column.split("!!")[0]

        value = enrichment_df[column][1]  
        if pd.notna(value):  
            state_data.append({"State": state_name, "Population Estimate": value})


transformed_df = pd.DataFrame(state_data)


print(transformed_df.head())
```

```python

merged_df = pd.merge(super_covid_df, transformed_df, left_on='State_x', right_on='State', how='left', suffixes=('_covid', '_pop'))


print(merged_df.head())

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

weekly_cases = pd.read_csv('./weekly_cases.csv')

week_df = pd.DataFrame(weekly_cases)

week_df
```

```python
us_df = week_df[['date', 'United States']]

us_df = us_df[us_df['date'] >= '2020-07-06']

us_df = us_df[us_df['date'] <= '2020-12-27']

us_df
```

```python
us_case_mean = us_df['United States'].mean()

us_case_mean
```

```python
us_case_mode = us_df.mode()

us_case_mode
```

```python
us_case_median = us_df['United States'].median()

us_case_median
```
